```python
fig = px.box(
    df,
    y='Revenue',
    title='Outlier Detection using Box Plot (Revenue)'
)
fig.show()
```

```python
revenue_col = df['Revenue']
Q1 = revenue_col.quantile(0.25)
Q3 = revenue_col.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print('Q1 =', Q1)
print('Q3 =', Q3)
print('IQR =', IQR)
print('Lower Limit =', lower_limit)
print('Upper Limit =', upper_limit)
outliers = df[(revenue_col < lower_limit) | (revenue_col > upper_limit)]
print('Number of outliers:', len(outliers))
df['outlier'] = (revenue_col < lower_limit) | (revenue_col > upper_limit)
df['Revenue'] = revenue_col.clip(lower_limit, upper_limit)
df.head()
```

#### No outliers were detected in the Revenue column using the IQR method, indicating that the data is well-distributed and does not contain extreme values that may distort the analysis.

```python
fig = px.box(
    df,
    y='Revenue',
    title='Outlier Detection using Box Plot (Revenue)'
)
fig.show()
```
